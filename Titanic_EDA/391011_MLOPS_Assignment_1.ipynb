```python
class TitanicEDA:
    def __init__(self, data_path):
        self.data = self.load_data(data_path)

    def load_data(self, data_path):
        return pd.read_csv(data_path)

    def generate_summary_statistics(self):
        """Generate summary statistics for numerical columns."""
        return self.data.describe()

    def visualize_survival_rate(self, feature, output_path):
        """Visualize survival rate based on a given feature."""
        plt.figure(figsize=(10, 6))
        sns.barplot(x=feature, y='Survived', data=self.data, errorbar=None, hue=feature, palette='viridis', legend=False)
        plt.title(f'Survival Rate by {feature}')
        plt.ylabel('Survival Rate')
        plt.savefig(output_path)
        plt.show()  # Show the plot
        plt.close()

    def visualize_age_distribution(self, output_path):
        """Visualize age distribution of passengers."""
        plt.figure(figsize=(10, 6))
        sns.histplot(data=self.data, x='Age', hue='Survived', multiple='stack', bins=20, palette='viridis')
        plt.title('Age Distribution of Passengers')
        plt.xlabel('Age')
        plt.ylabel('Count')
        plt.savefig(output_path)
        plt.show()  # Show the plot
        plt.close()

    def visualize_pclass_distribution(self, output_path):
        """Visualize passenger class distribution."""
        plt.figure(figsize=(10, 6))
        sns.countplot(data=self.data, x='Pclass', hue='Survived', palette='viridis')
        plt.title('Passenger Class Distribution')
        plt.xlabel('Passenger Class')
        plt.ylabel('Count')
        plt.savefig(output_path)
        plt.show()  # Show the plot
        plt.close()

    def run_eda(self):
        """Run the entire EDA process."""
        print("Summary Statistics:")
        print(self.generate_summary_statistics())

        self.visualize_survival_rate('Sex', 'survival_rate_by_sex.png')
        self.visualize_survival_rate('Pclass', 'survival_rate_by_pclass.png')
        self.visualize_age_distribution('age_distribution.png')
        self.visualize_pclass_distribution('pclass_distribution.png')

# Example usage:
# eda = TitanicEDA('titanic.csv')
# eda.run_eda()
```

```python
# Create an instance of TitanicEDA
eda = TitanicEDA('/content/drive/MyDrive/Titanic.csv')  # Replace with your actual file path

# Run the EDA process
eda.run_eda()
```
